```python
import pandas as pd
```

```python
df = pd.read_csv("cap_rooms.csv")
df
```

```python
df['PPN'] = df['PPN'].replace({'₹': '', ',': '', 'One Thousand Five Hundred': '1500', 'Two Thousand': '2000', 'Eight Hundred': '800', '₹': ''}, regex=True).astype(float)
```

```python
df['City'] = df['City'].replace({'Bangalor': 'Bangalore','bangaloreeee': 'bangalore', 'Hydrabad': 'Hyderabad'}, regex=True)
df
```

```python
df['Recent'] = pd.to_datetime(df['Recent'], errors='coerce')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['City'] = df['City'].replace({'Bangaloreeeee': 'Bangalore', 'Delhi ': 'Delhi'})
df
```

```python
df['Recent'] = df['Recent'].fillna(pd.to_datetime('2023-10-01'))
```

```python
df.to_csv('caps_updated_rooms.csv', index=False)
```
