## P5> Plot a graph showing the relationship between PPN and number of Bedrooms.

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("caps_updated_rooms.csv")
df
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Beds'], df['PPN'], alpha=0.6, edgecolors='w', s=100)
```

```python
plt.title('Relationship Between Price Per Night (PPN) and Number of Bedrooms', fontsize=16)
plt.xlabel('Number of Bedrooms', fontsize=14)
plt.ylabel('Price Per Night (PPN)', fontsize=14)
```

```python
plt.grid(True)
```

```python
plt.tight_layout()
plt.show()
```
